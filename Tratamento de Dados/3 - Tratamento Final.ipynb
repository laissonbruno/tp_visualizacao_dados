```python
import pandas as pd
latiLong = pd.read_csv('localizacao.csv', sep=',')
cancerData = pd.read_csv('CancerDataBase_Final.csv')

latiLong.info()
```

```python
# Criando duas listas para diferenciar colunas normais e colunas que contem dados do cancer.
colunas_Normais = ["Entity", "Code", "Year", "Region", "GDP per capita (current US$)", "Current health expenditure per capita (current US$)"]
colunas_Cancer = [col for col in cancerData.columns if col not in colunas_Normais]
```

```python
# Pivotando/remodelando o dataframe
cancerMelted = cancerData.melt(id_vars=colunas_Normais, value_vars=colunas_Cancer, var_name="Types", value_name="Deaths")
cancerMelted['Types'].unique()
```

```python
# def changeNames(old_name):
#     index = old_name[9:].find('-') + 8
#     return old_name[9:index]

# cancerMelted['Types'] = cancerMelted['Types'].apply(changeNames)
# cancerMelted['Types'].values

cancerMelted['Types'] = cancerMelted['Types'].str.replace('Deaths - ', '').str.replace(' - Sex: Both - Age: All Ages (Number)', '')
cancerMelted['Types'].unique()
```

```python
# depois tem que remover os dados nulos utilizando como base a coluna "Current health expenditure per capita (current US$)"
cancerMelted['Current health expenditure per capita (current US$)'].info()
# nenhuma entrada nula
```

```python
cancerMelted['Entity'].unique()
```

```python
# Limpando os dados nulos da coluna GDP
cancerMelted.dropna(subset=['GDP per capita (current US$)'], inplace=True)
cancerMelted.isnull().sum()
```

```python
cancerMelted['Entity'].unique()
```

```python
# Foi utilizado a ferramenta(https://www.geoapify.com/tools/geocoding-online), para a criação dos dados de localização dos paises, sendo eles a latitude e a longitude.
# Adicionando as colunas latitude e longitude 

cancerMelted_merged = cancerMelted.merge(latiLong, left_on='Entity', right_on='original_country', how='left')
cancerMelted_merged.drop(columns=['original_country'], inplace=True)
cancerMelted_merged.info()
```

```python
cancerMelted_merged['Entity'].unique()
```

```python
cancerMelted_merged.to_csv("CancerDataBase_Final_2.csv", index=False, mode='w')
```

Os dados aqui foram tratados para facilitar a visualização dos dados, ja que a disposição das colunas deixava com que a tabela ficasse grande demais horizontalmente o que dificulta a analise.

Também foi realizada a limpeza dos dados, retirando dados faltantes sobre gasto com saúde, nesse sentido também foi retirado os dados equivalentes sobre as mortes e demais dados equivalentes.
