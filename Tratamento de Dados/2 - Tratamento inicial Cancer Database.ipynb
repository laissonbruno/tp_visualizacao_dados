```
import pandas as pd
```

```
# Importando o arquivo com os dados sobre o câncer
cancerDataBase = pd.read_csv("total-cancer-deaths-by-type.csv")

# Importando o arquivo com os códigos ISO dos países
iso_country_code = pd.read_csv("iso-country-code.csv")

# Entradas a remover
retirar = pd.read_csv("retirar.csv")

# Importando o arquivo com os dados financeiros
financialDataBase = pd.read_csv("WDI_Tratado.csv")
```

```
#Filtrando dados do arquivo retirar.csv

print("original: "+str(cancerDataBase.shape[0]),"filtered: "+str(cancerDataBase.shape[0]-retirar.shape[0]))

df_merge = cancerDataBase.merge(retirar, on=['Entity', 'Year'], how='left', indicator=True)

DF_filtrado = cancerDataBase[df_merge['_merge'] == 'left_only']

print("after filter:", DF_filtrado.shape[0])
```

```
#Removendo dados anteriores a 1999
cancerDataBase = DF_filtrado[DF_filtrado['Year']>1999]
cancerDataBase.shape[0]
```

```
cancerDataBase['Year'].values
```

```
# Obtendo a lista de países únicos para a primeira verificação
paises_unicos_primeira_verificacao = cancerDataBase['Entity'].unique().tolist()
```

```
# Lista de nomes que serão removidos do dataframe
nomes = [
    "African Region (WHO)",
    "East Asia & Pacific (WB)",
    "Eastern Mediterranean Region (WHO)",
    "Europe & Central Asia (WB)",
    "European Region (WHO)",
    "Latin America & Caribbean (WB)",
    "Middle East & North Africa (WB)",
    "North America (WB)",
    "OECD Countries",
    "Region of the Americas (WHO)",
    "South Asia (WB)",
    "South-East Asia Region (WHO)",
    "Sub-Saharan Africa (WB)",
    "Western Pacific Region (WHO)",
    "World",
    "World Bank High Income",
    "World Bank Low Income",
    "World Bank Lower Middle Income",
    "World Bank Upper Middle Income"
]
```

```
# Mantém apenas as linhas onde 'Entity' não está na lista de nomes
cancerDataBase = cancerDataBase[~cancerDataBase['Entity'].isin(nomes)]
```

```
# Renomeia as colunas do dataframe iso_country_code e mescla com o dataframe cancerDataBase
iso_country_code = iso_country_code.rename(columns={'alpha-3': 'Code', 'region': 'Region'})
cancerDataBase = pd.merge(cancerDataBase, iso_country_code[['Code', 'Region']], on='Code', how='left')
```

```
# Seleciona as colunas necessárias do dataframe financialDataBase
financialDataBase = financialDataBase[['Country Code', 'Year', 'GDP per capita (current US$)', 'Current health expenditure per capita (current US$)']]
```

```
# Renomeia as colunas do dataframe financialDataBase para corresponder com o dataframe cancerDataBase
financialDataBase = financialDataBase.rename(columns={'Country Code': 'Code'})
```

```
# Mescla os dataframes cancerDataBase e financialDataBase
cancerDataBase = pd.merge(cancerDataBase, financialDataBase, on=['Code', 'Year'], how='left')
```

```
# Salva o DataFrame 'cancerDataBase' em um novo arquivo CSV chamado 'CancerDataBase_Final.csv'
cancerDataBase.to_csv("CancerDataBase_Final.csv", index=False)
```
