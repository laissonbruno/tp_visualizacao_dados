```
# Importa a biblioteca pandas
import pandas as pd
```

```
# Lê o arquivo CSV chamado 'WDICSV.csv' do diretório atual 
# e armazena os dados em um DataFrame do pandas chamado 'financialDataBase'
financialDataBase = pd.read_csv("WDICSV.csv")
```

```
# Cria uma lista chamada 'nomes' contendo os nomes dos indicadores de interesse
nomes = [
    "GDP per capita (current US$)",
    "Current health expenditure per capita (current US$)"
]
```

```
# Filtra o DataFrame 'financialDataBase' para incluir apenas as linhas onde o 'Indicator Name' está na lista 'nomes'
financialDataBase = financialDataBase[financialDataBase['Indicator Name'].isin(nomes)]
```

```
# Lista de colunas que não serão transpostas
id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
```

```
# Transpõe as colunas que contêm os anos, cria uma nova coluna 'Year' 
# e coloca os valores correspondentes nas novas colunas 'GDP per capita (current US$)' 
# e 'Current health expenditure per capita (current US$)'
financialDataBase = financialDataBase.melt(id_vars=id_vars, var_name='Year', value_name='Value')
```

```
# Cria um pivot table para separar os valores de 'GDP per capita (current US$)' e 'Current health expenditure per capita (current US$)' em colunas separadas
financialDataBase = financialDataBase.pivot_table(index=['Country Name', 'Country Code', 'Year'], columns='Indicator Name', values='Value').reset_index()
```

```
# Salva o DataFrame 'financialDataBase' em um novo arquivo CSV chamado 'WDI_Tratado.csv'
financialDataBase.to_csv("WDI_Tratado.csv", index=False)
```
